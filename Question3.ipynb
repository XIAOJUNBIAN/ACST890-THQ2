```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

<p><h1>Question a</h1></p>

```python
#read student_results.csv file
dataset = pd.read_csv('~/Downloads/student_results.csv')
dataset
```

<p><h1>Question b</h1></p>

```python
#replace NA to 0
dataset = dataset.fillna(0)
dataset
```

<p><h1>Question c</h1></p>

```python
#creat a weight martix
weight = np.array([[0.07*100/70],[0.03*100/30],[0.03*100/30],[0.02*100/20],[0.03*100/30],
                  [0.02*100/20],[0.024*100/30],[0.016*100/20],[0.07*100/70],[0.03*100/30],
                  [0.04*100/50],[0.02*100/2],[0.6*100/100]])
weight
```

```python
#use weight martix to calculate total.mark
dataset['Total.marks'] = np.dot(dataset.iloc[:,1:14],weight)
dataset
```

```python
#plot boxplot for different student categories
ax = sb.boxplot(x='Student category',y='Total.marks',data=dataset)
```

<p><h1>Question d</h1></p>

```python
#group the dataset by Class No and extract a number of stats from each group 
#find the mean, variance and standard deviation of the Total.marks column
stattable = dataset.groupby('Class No').agg({"Total.marks":['mean','var','std']})
stattable
```

<p><h1>Question e</h1></p>

```python
#creat grade for each student
Grade = []
for row in dataset['Total.marks']:
    if row < 50:
        Grade.append('F')
    elif row < 65:
        Grade.append('P')
    elif row <75:
        Grade.append('Cr')
    elif row <85:
        Grade.append('D')
    else:
        Grade.append('HD')
        
#creat a column called Grade
dataset['Grade'] = Grade
dataset
```

<p><h1>Question f</h1></p>

```python
import statsmodels.api as sm

y = dataset["Final exam"] #response
X = dataset["Assessment 1"] #predictor
X = sm.add_constant(X) #add a constant term to the predictor

X.head()
```

```python
#perform the regression of the predictor on the response

model = sm.OLS(y,X).fit()
predictions = model.predict(X) #make the predictions by the model

#print out the statistics
model.summary()
```

```python
#Plot the linear regression
fig, ax = plt.subplots()

#plot predicted values
plt.scatter(dataset["Assessment 1"], predictions, color='black', label='predicted')

#plot oberved values
plt.scatter(dataset["Assessment 1"], dataset["Final exam"],color='red', label='observed')

plt.legend()
plt.title('OLS predicted values')
plt.xlabel('Assessment 1')
plt.ylabel('Final exam')
plt.show()
```

```python
#Comment:
#For linear regression, coefficient of Assessment 1 is 0.2769, 
#which means Assessment1 increase 1 unit, Final exam increase 0.2769 unit
#intercept is 37.0586.
```

<p><h1>Question h</h1></p>

```python
#creat status for each students
Status =[]

for row in dataset['Grade']:
    if row == 'F':
        Status.append('incomplete')
    else:
        Status.append('complete')

#creat a column called Status
dataset['Status'] = Status
dataset
```

```python
#fit logistic regression model use training data of first 100 students
from sklearn.linear_model import LogisticRegression


#creat dummy variable for Status
y= pd.get_dummies(dataset.loc[0:99,'Status'])['complete']

X= dataset.loc[0:99,'Assessment 3'] #predictor
X= sm.add_constant(X) #add constant to predictor

logreg = LogisticRegression()
logreg.fit(X,y)
```

```python
from sklearn.metrics import confusion_matrix

#fit logistic for remaining data
y_test= pd.get_dummies(dataset.loc[100:,'Status'])['complete']
X_test= dataset.loc[100:,'Assessment 3'] #predictor
X_test= sm.add_constant(X_test) #add constant to predictor

y_pred= logreg.predict(X_test)

confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```
